```python
# To-Do List Uygulaması

import os
import sys

# Global değişken - görevler listesi
gorevler = []
```

```python

# DOSYA İŞLEMLERİ FONKSİYONLARI

def dosya_oku():
    """
    Görevleri dosyadan okur ve gorevler listesine yükler
    """
    global gorevler
    try:
        # Dosya var mı kontrol et
        if os.path.exists("gorevler.txt"):
            with open("gorevler.txt", "r", encoding="utf-8") as dosya:
                gorevler = []
                for satir in dosya:
                    # Boş satırları atla
                    satir = satir.strip()
                    if satir:
                        gorevler.append(satir)
            print(f"✅ {len(gorevler)} görev dosyadan yüklendi.")
        else:
            print("ℹ️ Henüz bir görev dosyası bulunamadı. Yeni bir liste oluşturuluyor.")
            gorevler = []
    except Exception as e:
        print(f"❌ Dosya okuma hatası: {e}")
        gorevler = []
```

```python
# DOSYA KAYDI
def dosya_kaydet():
    """
    Görevleri dosyaya kaydeder
    """
    try:
        with open("gorevler.txt", "w", encoding="utf-8") as dosya:
            for gorev in gorevler:
                dosya.write(gorev + "\n")
        print("💾 Görevler başarıyla kaydedildi.")
    except Exception as e:
        print(f"❌ Dosya kaydetme hatası: {e}")
```

```python

# GÖREV YÖNETİM FONKSİYONLARI

def gorevleri_listele():
    """
    Tüm görevleri numaralandırılmış şekilde listeler
    """
    print("\n" + "="*50)
    print("📋 GÖREV LİSTESİ")
    print("="*50)

    if not gorevler:
        print("📭 Henüz hiç görev eklenmemiş.")
    else:
        for i, gorev in enumerate(gorevler, 1):
            print(f"{i:2d}. {gorev}")

    print("="*50)

def yeni_gorev_ekle():
    """
    Yeni görev ekler
    """
    print("\n" + "="*50)
    print("➕ YENİ GÖREV EKLEME")
    print("="*50)

    while True:
        try:
            gorev = input("🆕 Yeni görev girin (iptal için 'q' yazın): ").strip()

            # İptal kontrolü
            if gorev.lower() == 'q':
                print("❌ Görev ekleme iptal edildi.")
                return

            # Boş görev kontrolü
            if not gorev:
                print("⚠️ Boş görev eklenemez! Lütfen geçerli bir görev girin.")
                continue

            # Görev ekleme
            gorevler.append(gorev)
            print(f"✅ '{gorev}' görevi başarıyla eklendi.")

            # Dosyaya kaydet
            dosya_kaydet()
            break

        except KeyboardInterrupt:
            print("\n❌ İşlem kullanıcı tarafından iptal edildi.")
            return
        except Exception as e:
            print(f"❌ Beklenmeyen hata: {e}")
```

```python
def gorev_sil():
    """
    Numaraya göre görev siler
    """
    print("\n" + "="*50)
    print("🗑️ GÖREV SİLME")
    print("="*50)

    # Önce görevleri göster
    if not gorevler:
        print("📭 Silinecek görev bulunmuyor.")
        return

    gorevleri_listele()

    while True:
        try:
            secim = input(f"\n🗑️ Silinecek görev numarası (1-{len(gorevler)}) [iptal: q]: ").strip()

            # İptal kontrolü
            if secim.lower() == 'q':
                print("❌ Görev silme iptal edildi.")
                return

            # Sayı kontrolü
            gorev_no = int(secim)

            # Geçerli aralık kontrolü
            if 1 <= gorev_no <= len(gorevler):
                silinecek_gorev = gorevler[gorev_no - 1]

                # Onay al
                onay = input(f"⚠️ '{silinecek_gorev}' görevi silinecek. Emin misiniz? (e/h): ").strip().lower()

                if onay == 'e':
                    gorevler.pop(gorev_no - 1)
                    print(f"✅ '{silinecek_gorev}' görevi başarıyla silindi.")
                    # Dosyaya kaydet
                    dosya_kaydet()
                    break
                else:
                    print("❌ Görev silme iptal edildi.")
                    return
            else:
                print(f"❌ Geçersiz görev numarası! 1-{len(gorevler)} arasında bir sayı girin.")

        except ValueError:
            print("❌ Geçersiz giriş! Lütfen bir sayı girin.")
        except KeyboardInterrupt:
            print("\n❌ İşlem kullanıcı tarafından iptal edildi.")
            return
        except Exception as e:
            print(f"❌ Beklenmeyen hata: {e}")
```

```python
def gorev_duzenle():
    """
    Numaraya göre görev düzenler
    """
    print("\n" + "="*50)
    print("✏️ GÖREV DÜZENLEME")
    print("="*50)

    # Önce görevleri göster
    if not gorevler:
        print("📭 Düzenlenecek görev bulunmuyor.")
        return

    gorevleri_listele()

    while True:
        try:
            secim = input(f"\n✏️ Düzenlenecek görev numarası (1-{len(gorevler)}) [iptal: q]: ").strip()

            # İptal kontrolü
            if secim.lower() == 'q':
                print("❌ Görev düzenleme iptal edildi.")
                return

            # Sayı kontrolü
            gorev_no = int(secim)

            # Geçerli aralık kontrolü
            if 1 <= gorev_no <= len(gorevler):
                eski_gorev = gorevler[gorev_no - 1]
                print(f"\n📝 Mevcut görev: '{eski_gorev}'")

                # Yeni görev al
                while True:
                    yeni_gorev = input("🆕 Yeni görev metni (iptal için 'q'): ").strip()

                    # İptal kontrolü
                    if yeni_gorev.lower() == 'q':
                        print("❌ Görev düzenleme iptal edildi.")
                        return

                    # Boş görev kontrolü
                    if not yeni_gorev:
                        print("⚠️ Boş görev olamaz! Lütfen geçerli bir görev girin.")
                        continue

                    # Görev güncelle
                    gorevler[gorev_no - 1] = yeni_gorev
                    print(f"✅ Görev başarıyla güncellendi:")
                    print(f"   Eski: '{eski_gorev}'")
                    print(f"   Yeni: '{yeni_gorev}'")

                    # Dosyaya kaydet
                    dosya_kaydet()
                    return

            else:
                print(f"❌ Geçersiz görev numarası! 1-{len(gorevler)} arasında bir sayı girin.")

        except ValueError:
            print("❌ Geçersiz giriş! Lütfen bir sayı girin.")
        except KeyboardInterrupt:
            print("\n❌ İşlem kullanıcı tarafından iptal edildi.")
            return
        except Exception as e:
            print(f"❌ Beklenmeyen hata: {e}")
```

```python

# MENÜ VE ANA PROGRAM

def menu_goster():

   # Ana menüyü gösterirg
    print("\n" + "="*50)
    print("📝 TO-DO LIST UYGULAMASI")
    print("="*50)
    print("1. 📋 Görevleri Listele")
    print("2. ➕ Yeni Görev Ekle")
    print("3. ✏️ Görev Düzenle")
    print("4. 🗑️ Görev Sil")
    print("5. 🚪 Çıkış")
    print("="*50)

def kullanici_secimi_al():
    """
    Kullanıcıdan menü seçimi alır ve geçerli seçim döndürür
    """
    while True:
        try:
            secim = input("👆 Seçiminizi yapın (1-5): ").strip()

            if secim in ['1', '2', '3', '4', '5']:
                return int(secim)
            else:
                print("❌ Geçersiz seçim! Lütfen 1-5 arasında bir sayı girin.")

        except KeyboardInterrupt:
            print("\n👋 Güle güle!")
            sys.exit()
        except Exception as e:
            print(f"❌ Beklenmeyen hata: {e}")

def ana_program():
    """
    Ana program döngüsü
    """
    print("🎯 TO-DO LIST UYGULAMASINA HOŞGELDİNİZ!")
    print("="*50)

    # Başlangıçta dosyadan görevleri oku
    dosya_oku()

    while True:
        try:
            # Menüyü göster
            menu_goster()

            # Kullanıcı seçimi al
            secim = kullanici_secimi_al()

            # Seçime göre işlem yap
            if secim == 1:
                gorevleri_listele()
            elif secim == 2:
                yeni_gorev_ekle()
            elif secim == 3:
                gorev_duzenle()
            elif secim == 4:
                gorev_sil()
            elif secim == 5:
                print("\n👋 Uygulamadan çıkılıyor...")
                print("💾 Tüm değişiklikler kaydedildi.")
                print("🙏 Güle güle!")
                break

        except KeyboardInterrupt:
            print("\n👋 Güle güle!")
            break
        except Exception as e:
            print(f"❌ Beklenmeyen hata: {e}")
            print("🔄 Program devam ediyor...")
```

```python
# ================================
# PROGRAM BAŞLATMA
# ================================

if __name__ == "__main__":
    ana_program()
```
