```python
# To-Do List UygulamasÄ±

import os
import sys

# Global deÄŸiÅŸken - gÃ¶revler listesi
gorevler = []
```

```python

# DOSYA Ä°ÅLEMLERÄ° FONKSÄ°YONLARI

def dosya_oku():
    """
    GÃ¶revleri dosyadan okur ve gorevler listesine yÃ¼kler
    """
    global gorevler
    try:
        # Dosya var mÄ± kontrol et
        if os.path.exists("gorevler.txt"):
            with open("gorevler.txt", "r", encoding="utf-8") as dosya:
                gorevler = []
                for satir in dosya:
                    # BoÅŸ satÄ±rlarÄ± atla
                    satir = satir.strip()
                    if satir:
                        gorevler.append(satir)
            print(f"âœ… {len(gorevler)} gÃ¶rev dosyadan yÃ¼klendi.")
        else:
            print("â„¹ï¸ HenÃ¼z bir gÃ¶rev dosyasÄ± bulunamadÄ±. Yeni bir liste oluÅŸturuluyor.")
            gorevler = []
    except Exception as e:
        print(f"âŒ Dosya okuma hatasÄ±: {e}")
        gorevler = []
```

```python
# DOSYA KAYDI
def dosya_kaydet():
    """
    GÃ¶revleri dosyaya kaydeder
    """
    try:
        with open("gorevler.txt", "w", encoding="utf-8") as dosya:
            for gorev in gorevler:
                dosya.write(gorev + "\n")
        print("ğŸ’¾ GÃ¶revler baÅŸarÄ±yla kaydedildi.")
    except Exception as e:
        print(f"âŒ Dosya kaydetme hatasÄ±: {e}")
```

```python

# GÃ–REV YÃ–NETÄ°M FONKSÄ°YONLARI

def gorevleri_listele():
    """
    TÃ¼m gÃ¶revleri numaralandÄ±rÄ±lmÄ±ÅŸ ÅŸekilde listeler
    """
    print("\n" + "="*50)
    print("ğŸ“‹ GÃ–REV LÄ°STESÄ°")
    print("="*50)

    if not gorevler:
        print("ğŸ“­ HenÃ¼z hiÃ§ gÃ¶rev eklenmemiÅŸ.")
    else:
        for i, gorev in enumerate(gorevler, 1):
            print(f"{i:2d}. {gorev}")

    print("="*50)

def yeni_gorev_ekle():
    """
    Yeni gÃ¶rev ekler
    """
    print("\n" + "="*50)
    print("â• YENÄ° GÃ–REV EKLEME")
    print("="*50)

    while True:
        try:
            gorev = input("ğŸ†• Yeni gÃ¶rev girin (iptal iÃ§in 'q' yazÄ±n): ").strip()

            # Ä°ptal kontrolÃ¼
            if gorev.lower() == 'q':
                print("âŒ GÃ¶rev ekleme iptal edildi.")
                return

            # BoÅŸ gÃ¶rev kontrolÃ¼
            if not gorev:
                print("âš ï¸ BoÅŸ gÃ¶rev eklenemez! LÃ¼tfen geÃ§erli bir gÃ¶rev girin.")
                continue

            # GÃ¶rev ekleme
            gorevler.append(gorev)
            print(f"âœ… '{gorev}' gÃ¶revi baÅŸarÄ±yla eklendi.")

            # Dosyaya kaydet
            dosya_kaydet()
            break

        except KeyboardInterrupt:
            print("\nâŒ Ä°ÅŸlem kullanÄ±cÄ± tarafÄ±ndan iptal edildi.")
            return
        except Exception as e:
            print(f"âŒ Beklenmeyen hata: {e}")
```

```python
def gorev_sil():
    """
    Numaraya gÃ¶re gÃ¶rev siler
    """
    print("\n" + "="*50)
    print("ğŸ—‘ï¸ GÃ–REV SÄ°LME")
    print("="*50)

    # Ã–nce gÃ¶revleri gÃ¶ster
    if not gorevler:
        print("ğŸ“­ Silinecek gÃ¶rev bulunmuyor.")
        return

    gorevleri_listele()

    while True:
        try:
            secim = input(f"\nğŸ—‘ï¸ Silinecek gÃ¶rev numarasÄ± (1-{len(gorevler)}) [iptal: q]: ").strip()

            # Ä°ptal kontrolÃ¼
            if secim.lower() == 'q':
                print("âŒ GÃ¶rev silme iptal edildi.")
                return

            # SayÄ± kontrolÃ¼
            gorev_no = int(secim)

            # GeÃ§erli aralÄ±k kontrolÃ¼
            if 1 <= gorev_no <= len(gorevler):
                silinecek_gorev = gorevler[gorev_no - 1]

                # Onay al
                onay = input(f"âš ï¸ '{silinecek_gorev}' gÃ¶revi silinecek. Emin misiniz? (e/h): ").strip().lower()

                if onay == 'e':
                    gorevler.pop(gorev_no - 1)
                    print(f"âœ… '{silinecek_gorev}' gÃ¶revi baÅŸarÄ±yla silindi.")
                    # Dosyaya kaydet
                    dosya_kaydet()
                    break
                else:
                    print("âŒ GÃ¶rev silme iptal edildi.")
                    return
            else:
                print(f"âŒ GeÃ§ersiz gÃ¶rev numarasÄ±! 1-{len(gorevler)} arasÄ±nda bir sayÄ± girin.")

        except ValueError:
            print("âŒ GeÃ§ersiz giriÅŸ! LÃ¼tfen bir sayÄ± girin.")
        except KeyboardInterrupt:
            print("\nâŒ Ä°ÅŸlem kullanÄ±cÄ± tarafÄ±ndan iptal edildi.")
            return
        except Exception as e:
            print(f"âŒ Beklenmeyen hata: {e}")
```

```python
def gorev_duzenle():
    """
    Numaraya gÃ¶re gÃ¶rev dÃ¼zenler
    """
    print("\n" + "="*50)
    print("âœï¸ GÃ–REV DÃœZENLEME")
    print("="*50)

    # Ã–nce gÃ¶revleri gÃ¶ster
    if not gorevler:
        print("ğŸ“­ DÃ¼zenlenecek gÃ¶rev bulunmuyor.")
        return

    gorevleri_listele()

    while True:
        try:
            secim = input(f"\nâœï¸ DÃ¼zenlenecek gÃ¶rev numarasÄ± (1-{len(gorevler)}) [iptal: q]: ").strip()

            # Ä°ptal kontrolÃ¼
            if secim.lower() == 'q':
                print("âŒ GÃ¶rev dÃ¼zenleme iptal edildi.")
                return

            # SayÄ± kontrolÃ¼
            gorev_no = int(secim)

            # GeÃ§erli aralÄ±k kontrolÃ¼
            if 1 <= gorev_no <= len(gorevler):
                eski_gorev = gorevler[gorev_no - 1]
                print(f"\nğŸ“ Mevcut gÃ¶rev: '{eski_gorev}'")

                # Yeni gÃ¶rev al
                while True:
                    yeni_gorev = input("ğŸ†• Yeni gÃ¶rev metni (iptal iÃ§in 'q'): ").strip()

                    # Ä°ptal kontrolÃ¼
                    if yeni_gorev.lower() == 'q':
                        print("âŒ GÃ¶rev dÃ¼zenleme iptal edildi.")
                        return

                    # BoÅŸ gÃ¶rev kontrolÃ¼
                    if not yeni_gorev:
                        print("âš ï¸ BoÅŸ gÃ¶rev olamaz! LÃ¼tfen geÃ§erli bir gÃ¶rev girin.")
                        continue

                    # GÃ¶rev gÃ¼ncelle
                    gorevler[gorev_no - 1] = yeni_gorev
                    print(f"âœ… GÃ¶rev baÅŸarÄ±yla gÃ¼ncellendi:")
                    print(f"   Eski: '{eski_gorev}'")
                    print(f"   Yeni: '{yeni_gorev}'")

                    # Dosyaya kaydet
                    dosya_kaydet()
                    return

            else:
                print(f"âŒ GeÃ§ersiz gÃ¶rev numarasÄ±! 1-{len(gorevler)} arasÄ±nda bir sayÄ± girin.")

        except ValueError:
            print("âŒ GeÃ§ersiz giriÅŸ! LÃ¼tfen bir sayÄ± girin.")
        except KeyboardInterrupt:
            print("\nâŒ Ä°ÅŸlem kullanÄ±cÄ± tarafÄ±ndan iptal edildi.")
            return
        except Exception as e:
            print(f"âŒ Beklenmeyen hata: {e}")
```

```python

# MENÃœ VE ANA PROGRAM

def menu_goster():

   # Ana menÃ¼yÃ¼ gÃ¶sterirg
    print("\n" + "="*50)
    print("ğŸ“ TO-DO LIST UYGULAMASI")
    print("="*50)
    print("1. ğŸ“‹ GÃ¶revleri Listele")
    print("2. â• Yeni GÃ¶rev Ekle")
    print("3. âœï¸ GÃ¶rev DÃ¼zenle")
    print("4. ğŸ—‘ï¸ GÃ¶rev Sil")
    print("5. ğŸšª Ã‡Ä±kÄ±ÅŸ")
    print("="*50)

def kullanici_secimi_al():
    """
    KullanÄ±cÄ±dan menÃ¼ seÃ§imi alÄ±r ve geÃ§erli seÃ§im dÃ¶ndÃ¼rÃ¼r
    """
    while True:
        try:
            secim = input("ğŸ‘† SeÃ§iminizi yapÄ±n (1-5): ").strip()

            if secim in ['1', '2', '3', '4', '5']:
                return int(secim)
            else:
                print("âŒ GeÃ§ersiz seÃ§im! LÃ¼tfen 1-5 arasÄ±nda bir sayÄ± girin.")

        except KeyboardInterrupt:
            print("\nğŸ‘‹ GÃ¼le gÃ¼le!")
            sys.exit()
        except Exception as e:
            print(f"âŒ Beklenmeyen hata: {e}")

def ana_program():
    """
    Ana program dÃ¶ngÃ¼sÃ¼
    """
    print("ğŸ¯ TO-DO LIST UYGULAMASINA HOÅGELDÄ°NÄ°Z!")
    print("="*50)

    # BaÅŸlangÄ±Ã§ta dosyadan gÃ¶revleri oku
    dosya_oku()

    while True:
        try:
            # MenÃ¼yÃ¼ gÃ¶ster
            menu_goster()

            # KullanÄ±cÄ± seÃ§imi al
            secim = kullanici_secimi_al()

            # SeÃ§ime gÃ¶re iÅŸlem yap
            if secim == 1:
                gorevleri_listele()
            elif secim == 2:
                yeni_gorev_ekle()
            elif secim == 3:
                gorev_duzenle()
            elif secim == 4:
                gorev_sil()
            elif secim == 5:
                print("\nğŸ‘‹ Uygulamadan Ã§Ä±kÄ±lÄ±yor...")
                print("ğŸ’¾ TÃ¼m deÄŸiÅŸiklikler kaydedildi.")
                print("ğŸ™ GÃ¼le gÃ¼le!")
                break

        except KeyboardInterrupt:
            print("\nğŸ‘‹ GÃ¼le gÃ¼le!")
            break
        except Exception as e:
            print(f"âŒ Beklenmeyen hata: {e}")
            print("ğŸ”„ Program devam ediyor...")
```

```python
# ================================
# PROGRAM BAÅLATMA
# ================================

if __name__ == "__main__":
    ana_program()
```
